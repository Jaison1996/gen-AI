```python
import numpy as np

n=50
df=pd.DataFrame({
    'emp_id': range(1000, 1000+n),
    'name': [f'Employee_{i}' for i in range(n)],
    'age': np.random.randint(20, 60, size=n),
    'department': np.random.choice(['Engineering', 'HR', 'Sales'], n),
     'rating':np.random.choice([1,2,3,4,5],n),
     'salary': np.random.randint(50000, 150000, size=n)
    
})
df
```

```python
# Dataset with intentional missing values
data = {
   'Name'  : ['Alice','Bob','Carol','Dave','Eve','Frank'],
   'Age'   : [24, np.nan, 27, 35, np.nan, 40],
   'Salary': [72000, 55000, np.nan, 62000, 48000, np.nan],
   'Dept'  : ['Eng', 'Mkt', 'Eng', None, 'Mkt', 'Eng'],
   'Score' : [85, 90, np.nan, 78, 82, np.nan],
}
df = pd.DataFrame(data)
print("Dataset with missing values:\n", df)
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df['Salary'].mean()
```

```python
df_filled = df.copy()
```

```python
df_filled['Age']    = df_filled['Age'].fillna(df_filled['Age'].median())
df_filled['Salary'] = df_filled['Salary'].fillna(df_filled['Salary'].mean())
df_filled['Dept']   = df_filled['Dept'].fillna('Unknown')

df_filled
```

```python
df_filled.groupby('Dept')['Salary'].sum()
```

```python
summary = df.groupby('Dept').agg(
   Count_Emp      = ('Name', 'count'),
   Mean_Salary = ('Salary', 'mean'),
   Max_Score = ('Score', 'max'),
   Avg_Salary  = ('Salary',  'mean'),
).round(1)
print(summary)
```

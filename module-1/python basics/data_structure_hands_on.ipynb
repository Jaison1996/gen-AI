```python
#LIST
#Ordered, Mutable, allow duplicates

vechicles = ['bike', 'bus', 'car']
print(vechicles)
 
vechicles.append('cycle')
print(vechicles)
 
vechicles.remove('bike')
print(vechicles)
 
print(vechicles[0])
vechicles[0] = 'cycle'
print(vechicles)
print(len(vechicles))
```

```python
# Tuple
# Ordered, Immutable, Allows Duplicates
# To have secured information
# Call history, bank transaction
# GPS - Coordinate - Place
chennai_coords = (13.0823, 80.2707)
print(chennai_coords)
print('Latitue', chennai_coords[0])
print(len(chennai_coords))
```

```python
# SET
# Unordered, Mutable, No Duplicates
courses = { 'python', 'gen-ai', 'java', 'python' }
print(courses)
```

```python
# Dictionary
# Key-Value Pair, Unordered, Mutable, No Duplicates
# Depart name, employee records
# Student mark
 
report_card = {
    'name' : 'Arjun',
    'maths' : 90,
    'science' : 80,
    'english' : 100
}
 
print(report_card)
report_card['maths'] = 95
print(report_card)
```

```python
# Array
# Ordered, Mutable, Allows Duplicates, Homogeneous Data
import array
 
temp = array.array('f', [28.5, 30.1, 27.8, 34.0])
print(temp)
temp.append(34.0)
print(temp)
tmp = list(temp)
tmp.append(90)
temp = array.array('f',tmp)
print(temp)
```

```python
# String
# Ordered, Immutable, Allows Duplicates
ticket = 'aiAI-12345'
ticket = ticket + '678'
print(ticket)
 
print(ticket.upper())
print(ticket.split('-'))
print(ticket. Replace('aiAI', 'AI'))
```

```python
# Stack
# LIFO - Last In First Out
 
# Browser history
# Bank transaction
browser_history = []
browser_history.append('google.com')
browser_history.append('github.com')
 
 
print(browser_history)
# pop
last_one = browser_history.pop()
print(last_one)
print(browser_history)
```

```python
# Queue
# Ticket booking system
# FIFO
# Order processing
# Online message -> SMS system ->
# Avoid outage -> Slowness -> 8GB (100) -> 99
# in Queue -> 0 1
from collections import deque
ticket_queue = deque()
 
ticket_queue.append('Rahul')
ticket_queue.append('Uma')
ticket_queue.append('Jaison')
 
served = ticket_queue.popleft()
print(served)
print(ticket_queue)
served = ticket_queue.popleft()
print(served)
```

```python
# Named tuple
# Employee record - name, id, department
from collections import namedtuple
 
FingerPrint = namedtuple('FinderPrintTransaction', ['emp_id', 'time'])
emp1 = FingerPrint(emp_id='Emp01', time='09.00')
emp2 = FingerPrint(emp_id='Emp02', time='09.01')
 
print(emp1)
 
coll = [emp1, emp2]
print(coll)
```
